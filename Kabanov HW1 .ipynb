```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
s=10                  #10
r=28                 #28
b=2.667               #2.667
```

```python
def lorenz(xyz,t):
    x,y,z=xyz
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return x_dot, y_dot, z_dot
```

```python
initial = (-10, -7, 35)
t = np.arange(0, 100, 0.006)

solution = odeint(lorenz, initial, t)

xs = solution[:, 0]
ys = solution[:, 1]
zs = solution[:, 2]
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.show()
```
