```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
s=10                  #10
r=28                 #28
b=2.667               #2.667
```

```python
def lorenz(xyz,t):
    x,y,z=xyz
    xd = s*(y - x)
    yd = x*(r - z) - y 
    zd = x*y - b*z
    return xd, yd, zd
```

Аттрактор Лоренса
$\left\{
    \begin{gathered}
        \dot{x} = \sigma*(y-x)\\ 
        \dot{y} = x*(r-z)-y\\ 
        \dot{z} = x*y - b*z
    \end{gathered} 
\right.$

```python
initial = (-10, -10, 10)
t = np.arange(0, 100, 0.006)

solution = odeint(lorenz, initial, t)

xs = solution[:, 0]
ys = solution[:, 1]
zs = solution[:, 2]
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.show()
```
